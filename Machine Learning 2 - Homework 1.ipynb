##### 1. Divide the data in two groups: training and test examples.

```python
import os 
import numpy as np 
from collections import Counter 
from sklearn.metrics import confusion_matrix
```

```python
data = [] 
with open('messages.txt') as m: 
        data += m
        
data_train = data[0:int(len(data)*0.8)]
data_test = data[int(len(data)*0.8):len(data)]

print("longueur de data_train: ", len(data_train))
print("longueur de data_test: ", len(data_test))
```

##### 2. Parse both the training and test examples to generate both the spam and ham data sets.

##### 3. Generate a dictionary from the training data.

```python
data_train
```

```python
import os 
import numpy as np 
from sklearn.metrics import confusion_matrix

# Cleaning data train
data_train_cl = []
for i in data_train:
    data_train_cl.append(i[4:len(i)-1])

all_words = [] 
for j in data_train_cl:
    words = j.split() # split en fonction d'un espace
    all_words += words
dictionary = Counter(all_words)
# print(dictionary)
print("longeur du dictionnaire: ",len(dictionary))

for item in list(dictionary): 
    #check si item est bien un string contenant que des alphabets
    if item.isalpha() == False: 
        del dictionary[item] 
    elif len(item) == 1: 
        del dictionary[item] 
# garde les 3000 mots les plus communs (les 3000 mots qui ont le + d'occurence)
dictionary = dictionary.most_common(3000)
print("longeur du dictionnaire: ",len(dictionary))
print(dictionary)
```

##### 4. Extract features from both the training data and test data.

```python
# Cleaning data test
data_test_cl = []
for i in data_test:
    data_test_cl.append(i[4:len(i)-1])
data_test_cl
```

```python
def extract_features(data): 
    features_matrix = np.zeros((len(data), 3000))
    docID = 0
    for mail in data:
        words = mail.split() 
        for word in words: 
            wordID = 0 
            for i, d in enumerate(dictionary): 
                if d[0] == word: 
                    wordID = i 
                    features_matrix[docID, wordID] = words.count(word) 
        docID = docID + 1 
    return features_matrix
```

```python
# Extract features from training data.
train_matrix = extract_features(data_train_cl)
print('taille de la matrice : ', train_matrix.shape)
train_matrix
```

```python
# Extract features from both the test data.
test_matrix = extract_features(data_test_cl)
print('taille de la matrice : ', test_matrix.shape)
test_matrix
```

```python
# Cleaning spam data train
spam_data_train_cl = []
for i in spam_data_train:
    spam_data_train_cl.append(i[4:len(i)-1])
    
spam_train_matrix = extract_features(spam_data_train_cl)
print('taille de la matrice : ', spam_train_matrix.shape)
spam_train_matrix
```

```python
# Cleaning ham data train
ham_data_train_cl = []
for i in ham_data_train:
    ham_data_train_cl.append(i[4:len(i)-1])
    
ham_train_matrix = extract_features(ham_data_train_cl)
print('taille de la matrice : ', ham_train_matrix.shape)
ham_train_matrix
```

##### 5. Implement the Naive Bayes from scratch, ﬁt the respective models to the training data.

```python
alpha = 1

# φy = P(y(i) = 1) :
p_y_1 = (y_train.count('1') + alpha) / (len(y_train) + 2*alpha)
# p_y_0 = (y_train.count('0') + alpha) / (len(y_train) + 2*alpha)
p_y_0 = 1 - p_y_1


# φn|y=1 = P(xn = 1|y = 1) :
v1 =np.zeros((3000))
for j in range(spam_train_matrix.shape[1]):
    for i in range(spam_train_matrix.shape[0]):
        v1[j] += spam_train_matrix[i][j]

p_x_1_y_1 = (v1 + alpha) / (y_train.count('1') + 2*alpha)

# φn|y=0 = P(xn = 1|y = 0) :
v0 =np.zeros((3000))
for j in range(ham_train_matrix.shape[1]):
    for i in range(ham_train_matrix.shape[0]):
        v0[j] += ham_train_matrix[i][j]
    
p_x_1_y_0 = (v0 + alpha) / (y_train.count('0') + 2*alpha)

print('φy = P(y(i) = 1) =', p_y_1,'\n')
print('φy = P(y(i) = 0) =', p_y_0,'\n')

print('φn|y=1 = P(xn = 1|y = 1) = ', p_x_1_y_1,'\n')
print('φn|y=0 = P(xn = 1|y = 0) = ', p_x_1_y_0)
```

```python
# prediction_spam with Laplace smoothing technique :
proba_spam = (p_x_1_y_1 * p_y_1) / ((p_x_1_y_1 * p_y_1) + (p_x_1_y_0 * p_y_0))
# myround = np.vectorize(round)
print(proba_spam)

# prediction_ham with Laplace smoothing technique :
proba_ham = (p_x_1_y_0 * p_y_0) / ((p_x_1_y_1 * p_y_1) + (p_x_1_y_0 * p_y_0))
print(proba_ham)
```

```python
target_spam = np.zeros(len(spam_train_matrix))
for email in spam_train_matrix:
    for i, word, proba in zip(range(len(spam_train_matrix)), email, proba_spam):
        if word != 0:
            if target_spam[i] == 0:
                target_spam[i] = proba
            else:
                target_spam[i] *= proba
print(len(target_spam))
target_spam
```

```python
target_ham = np.zeros(len(spam_train_matrix))
for email in spam_train_matrix:
    for i, word, proba in zip(range(len(spam_train_matrix)), email, proba_ham):
        if word != 0:
            if target_ham[i] == 0:
                target_ham[i] = proba
            else:
                target_ham[i] *= proba
print(len(target_ham))
target_ham
```

```python
TARGET = []
for i in (target_spam-target_ham):
    if i > 0:
        TARGET.append(1)
    else:
        TARGET.append(0)
TARGET
```

```python
accuracy = 0
for target, y in zip(TARGET, y_train):
    if target == int(y):
        accuracy += 1
accuracy = accuracy / len(TARGET)
print(accuracy)
```
