### 1. Divide the data in two groups: training and test examples.

```python
import os 
import numpy as np 
from collections import Counter 
```

```python
data = [] 
with open('messages.txt') as m: 
        data += m
        
data_train = data[0:int(len(data)*0.8)]
data_test = data[int(len(data)*0.8):len(data)]

print("longueur de data_train: ", len(data_train))
print("longueur de data_test: ", len(data_test))
```

### 2. Parse both the training and test examples to generate both the spam and ham data sets.

```python
# spam : 1 & ham : 0
y_test = [] 

for i in data_test:
    if i.startswith('spam'):
        y_test.append(1)
    else:
        y_test.append(0)
        
y_train = []

for i in data_train:
    if i.startswith('spam'):
        y_train.append(1)
    else:
        y_train.append(0)
```

```python
print("longueur de y_train: ", len(y_train))
print(Counter(y_train))
```

```python
print("longueur de y_test: ", len(y_test))
print(Counter(y_test))
```

### 3. Generate a dictionary from the training data.

```python
data_train
```

##### Cleaning data_train

* Removing the "spam"/"ham", "\t" and "\n"

```python
data_train_cl = []
for i,msg in zip(range(len(data_train)),data_train):
    if y_train[i] == 1:
        data_train_cl.append(msg[5:len(msg)-1])
    else:
        data_train_cl.append(msg[4:len(msg)-1])

data_train_cl
```

* Creating a list of all words with their occurences

```python
all_words = [] 

for j in data_train_cl:
    words = j.split() # split en fonction d'un espace
    all_words += words
dictionary = Counter(all_words)
print(dictionary)
#print("Number of words: ",len(dictionary))

```

* Removing the non-alphabetic and 1-letter words

```python
for item in list(dictionary): 
    if item.isalpha() == False: 
        del dictionary[item] 
    elif len(item) == 1: 
        del dictionary[item]

print(dictionary)
```

* Keeping only the 3 000 most common words

```python
dictionary = dictionary.most_common(3000)
print("Dictionnary's length: ",len(dictionary))
print(dictionary)
```

### 4. Extract features from both the training data and test data.

* Removing the "spam"/"ham", "\t" and "\n"

```python
# Cleaning data test
data_test_cl = []
for i,msg in zip(range(len(data_test)),data_test):
    if y_test[i] == 1:
        data_test_cl.append(msg[5:len(msg)-1])
    else:
        data_test_cl.append(msg[4:len(msg)-1])

data_test_cl
```

```python
def extract_features(data): 
    features_matrix = np.zeros((len(data), 3000))
    docID = 0
    for mail in data:
        words = mail.split() 
        for word in words: 
            wordID = 0 
            for i, d in enumerate(dictionary): 
                if d[0] == word: 
                    wordID = i 
                    features_matrix[docID, wordID] = words.count(word) 
        docID = docID + 1 
    return features_matrix
```

* Extract features from training data.


```python
x_train = extract_features(data_train_cl)
print('x_train shape : ', x_train.shape)
x_train
```

* Extract features from test data.


```python
x_test = extract_features(data_test_cl)
print('x_test shape : ', x_test.shape)
x_test
```

### 5. Implement the Naive Bayes from scratch, ﬁt the respective models to the training data.

#### Definition of α (Laplace smoothing technique variable)  

```python
alpha = 0.06
D = len(dictionary)
```

#### P(y = 1)

```python
# I = number of messages
I = len(y_train)

# indicator function for {y=1}
I_y1 = y_train.count(1)

P_y1 = (I_y1 + alpha)/(I + 2*alpha)

print(' P(y=1) = ',P_y1)
```

#### P(y = 0)

```python
# indicator function for {y=0}
I_y0 = y_train.count(0)

P_y0 = 1 - P_y1

print("P(y=0) = ", P_y0)
```

#### P(xn = 1|y = 1)


```python
# Indicator function for {xn=1,y=1}
I_x1_y1 = np.zeros(len(dictionary))

# Creation of a matrix with both y and x 
y_train_np = np.reshape(y_train, (4000,1))

yx_train = np.concatenate((y_train_np,x_train), axis=1)

# Calculation of the indicator function
for col in range(1,yx_train.shape[1]):
    for li in range(yx_train.shape[0]):
        if yx_train[li][0] == 1:    # if y=1
            I_x1_y1[col-1] += yx_train[li][col]
            
phi_y1 = (I_x1_y1 + alpha) / (I_y1 + alpha * D)

phi_y1
```

#### P(xn = 1|y = 0)


```python
# Indicator function for {xn=1,y=0}
I_x1_y0 = np.zeros(len(dictionary))

for col in range(1,yx_train.shape[1]):
    for li in range(yx_train.shape[0]):
        if yx_train[li][0] == 0:
            I_x1_y0[col-1] += yx_train[li][col]
            
phi_y0 = (I_x1_y0 + alpha) / (I_y0 + alpha * D)

phi_y0
```

#### P(x|y = 1)


```python
spam_train = np.zeros(y_train_np.shape[0])

for li in range(x_train.shape[0]):
    index = []
    prod = 1
    for col in range(x_train.shape[1]):
        if x_train[li][col] != 0:
            index.append(col)
    for i in index:
        prod *= phi_y1[i]
        
    spam_train[li] = prod

spam_train
```

#### P(x|y = 0)


```python
ham_train = np.zeros(y_train_np.shape[0])

for li in range(x_train.shape[0]):
    index = []
    prod = 1
    for col in range(x_train.shape[1]):
        if x_train[li][col] != 0:
            index.append(col)
    for i in index:
        prod *= phi_y0[i]
        
    ham_train[li] = prod

ham_train
```

#### P(y = 1|x)

```python
P_y1_x = (spam_train * P_y1) / ( spam_train * P_y1 + ham_train * P_y0 )

P_y1_x
```

#### P(y = 0|x)


```python
P_y0_x = (ham_train * P_y0) / ( spam_train * P_y1 + (ham_train * P_y0))

P_y0_x
```

```python
target_train = []

for i in range(len(P_y1_x)):
    if P_y1_x[i] < P_y0_x[i]:
        target_train.append(0)
    else: 
        target_train.append(1)
    
target_train
```

### 6. Make predictions for the test data.

#### P(x|y = 1)


```python
spam_test = np.zeros(x_test.shape[0])

for li in range(x_test.shape[0]):
    index = []
    prod = 1
    for col in range(x_test.shape[1]):
        if x_test[li][col] != 0:
            index.append(col)
    for i in index:
        prod *= phi_y1[i]
        
    spam_test[li] = prod

spam_test
```

```python
ham_test = np.zeros(x_test.shape[0])

for li in range(x_test.shape[0]):
    index = []
    prod = 1
    for col in range(x_test.shape[1]):
        if x_test[li][col] != 0:
            index.append(col)
    for i in index:
        prod *= phi_y0[i]
        
    ham_test[li] = prod

ham_test
```

#### P(y = 1|x)

```python
P_y1_x_test = (spam_test * P_y1) / ( spam_test * P_y1 + (ham_test * P_y0))

P_y1_x_test
```

#### P(y = 0|x)


```python

P_y0_x_test = (ham_test * P_y0) / ( spam_test * P_y1 + (ham_test * P_y0))

P_y0_x_test
```

```python
target_test = []

for i in range(len(P_y1_x_test)):
    if P_y1_x_test[i] < P_y0_x_test[i]:
        target_test.append(0)
    else: 
        target_test.append(1)
    
target_test
```

### 7. Measure the spam-ﬁltering performance for each approach through the confusion matrix.

```python
def myConfusionMatrix(predicted,actual):
    
    TP = 0
    FP = 0
    FN = 0
    TN = 0

    for i in range(len(predicted)):
        if predicted[i] == 1 and int(actual[i]) == 1:
            TP += 1
        elif predicted[i] == 1 and int(actual[i]) == 0:
            FP += 1
        elif predicted[i] == 0 and int(actual[i]) == 1:
            FN += 1
        else:
            TN += 1
            
    Accuracy = (TP+TN)/(TP+FP+FN+TN)
    Precision = TP/(TP+FP)
    Recall = TP/(TP+FN)
    F1Score = 2*(Recall * Precision) / (Recall + Precision)
    
    print(['TP','FP'])
    print(['FN','TN'], '\n')
    print(np.array([[TP, FP],[FN, TN]]), '\n')
    
    print('Accuracy = ',Accuracy ,
          '\nPrecision = ',Precision,
          '\nRecall = ',Recall,
          '\nF1 Score = ',F1Score
         )
    return np.array([[TP, FP],[FN, TN]])
```
